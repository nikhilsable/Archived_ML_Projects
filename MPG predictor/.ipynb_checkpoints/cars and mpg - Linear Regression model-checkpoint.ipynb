```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('auto.csv')
```

```python
#Input data had a "?" for missing values in HP column. I could've rectified this during import itself (na_values = '?'),
#but I only learnt this after I ran this clean-up. Keeping it here as a lesson - if you can, you should!
data['horsepower'].str.contains('?', regex = False)
```

```python
data = data[data['horsepower'].str.contains('?', regex = False) != True]

#I had to clean the data a little bit since the 'horsepower' feature/column had some bad data
# A big thanks to UCI for providing the dataset - http://archive.ics.uci.edu/ml/datasets/Auto+MPG 
```

```python
data.head(10) #just to make sure
```

```python
data['horsepower'] = data['horsepower'].astype(float)
```

```python
data.dtypes
```

```python
X = data[['cylinders', 'displacement', 'horsepower', 'weight','acceleration','model year', 'origin']]
```

```python
#I like ot see data. Plus Seaborn is awesome, so why not. 

import seaborn as sns
sns.pairplot(data = data, x_vars=['cylinders', 'displacement','horsepower','weight','acceleration','model year', 'origin'], y_vars=['mpg'], size = 5, aspect = 0.4, kind='reg')
```

```python
#let's verify our datatype for y/dependant variable/response is series
y = data['mpg']
type(y)
```

### Model #1

```python
list(X)
```

```python
X = data[['cylinders', 'displacement', 'horsepower', 'weight','acceleration','model year', 'origin']]
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn import metrics

#instantiate (instance) of a linear reg model
linreg = LinearRegression()

#split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 1)

#fit
linreg.fit(X_train,y_train)

#predict
y_pred = linreg.predict(X_test)

#evaluate using RMSE (there are other methods to eval too - MSE, etc)
print (np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#linreg(intercept_)
print (linreg.intercept_)
print(linreg.coef_)
```

##### Now that we have a error value, let's see if we can better the model by playing with the features we've used in the model

### I'm not a big fan of "origin" and "cylinders" (based on the graph above), so let's start by removing origin

### Model 1 wins!! 

Model 1 seems to have the lowest RMSE and hence seems to be a good fit. Using Model 1 we can effectively gauge the association (not causation) of features in the data set t othe response/dependant variable (mpg).
